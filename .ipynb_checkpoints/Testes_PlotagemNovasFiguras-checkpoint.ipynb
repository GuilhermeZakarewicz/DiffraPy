```python
from obspy import read, Trace, Stream, UTCDateTime
from obspy.core import AttribDict
from obspy.io.segy.segy import SEGYTraceHeader, SEGYBinaryFileHeader
from obspy.io.segy.core import _read_segy, _read_su
import numpy as np
import sys
import matplotlib.pyplot as plt
import Diffrapy as df
```

```python
filename = "Data/SEGY_files/teste_sudiffsummtoolow.su"
filename2 = "Data/SEGY_files/teste_sudiffsummtoohigh.su"

st = _read_su(filename)
st2 = _read_su(filename2)
#st_o = _read_su(filename_original)
dt = st[0].stats.delta
dx = 10
dz = dx
nx = np.shape(st)[0]
nt = np.shape(st)[1]

gather = np.zeros([nx,nt])
#gather_o = np.copy(gather)
gather2 = np.copy(gather)

for i in range(len(gather)):
    gather[i] = st[i]
    #gather_o[i] = st_o[i]
    gather2[i] = st2[i]

teste_diffs = gather.T
teste2_diffs = gather2.T
#original_diffs = gather_o.T

plt.imshow(teste_diffs, aspect="auto", cmap="binary")
plt.show()

plt.imshow(teste2_diffs, aspect="auto", cmap="binary")
plt.show()

#plt.imshow(original_diffs, aspect="auto", cmap="binary")
#plt.show()
```

```python
filename = "Data/SEGY_files/teste_sudiffhomo_tpow1_filter3035110115.su"
filename2 = "Data/SEGY_files/teste_sudiffhomo_tpow1gpow07_f3035110115.su"
filename_original = "Data/SEGY_files/sudiffs_homo.su"

st = _read_su(filename)
st2 = st = _read_su(filename2)
st_o = _read_su(filename_original)
dt = st[0].stats.delta
dx = 10
dz = dx
nx = np.shape(st)[0]
nt = np.shape(st)[1]

gather = np.zeros([nx,nt])
gather_o = np.copy(gather)
gather2 = np.copy(gather)

for i in range(len(gather)):
    gather[i] = st[i]
    gather_o[i] = st_o[i]
    gather2[i] = st2[i]

teste_diffs = gather.T
teste2_diffs = gather2.T
original_diffs = gather_o.T

plt.imshow(teste_diffs, aspect="auto", cmap="binary")
plt.show()

plt.imshow(teste2_diffs, aspect="auto", cmap="binary")
plt.show()

plt.imshow(original_diffs, aspect="auto", cmap="binary")
plt.show()
```
