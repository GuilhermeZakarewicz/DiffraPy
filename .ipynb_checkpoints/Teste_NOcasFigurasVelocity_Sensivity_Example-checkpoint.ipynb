```python
import importlib
import numpy as np
import matplotlib.pyplot as plt
import Diffrapy as df
```

```python
nx=150 #meters
nz=100 #meters
ntr=150 #number of traces in acquisition
dx=1.0 #meter
dz=1.0 #meter
c=250 #constant velocity [m/s]
dt=.004 #seconds
```

```python
vel1_toohigh = np.ones([nz,ntr])*c*1.05
vel1_toolow = np.ones([nz,ntr])*c*0.95
SW_toohigh = 1./vel1_toohigh
SW_toolow = 1./vel1_toolow
```

```python
#TT_toohigh = df.raymodel3(SW_toohigh,dx,ntr,"Data/CSG_diffbar/raiosHomo_toohigh.npy")
#TT_toolow = df.raymodel3(SW_toolow,dx,ntr,"Data/CSG_diffbar/raiosHomo_toolow.npy")
```

```python
dsx = 15 #Shots spacing
nsx = nx
ngx = nx

files = []

for i in range(0,nsx,dsx):
    filename = 'Data/CSG_diffbar/Toy_model_gathershot_{}'.format(i)
    with open(filename, 'rb') as file:
        (gather) = np.load(file)
        files.append(gather)
```

```python
with open("Data/CSG_diffbar/Homogeneous_model_pm", "rb") as f:
    pm = np.load(f)

with open("Data/CSG_diffbar/Homogeneous_model_semblance", "rb") as f:
    sm = np.load(f)
```

```python
m_theta = np.arctan(pm)
X = np.sin(m_theta)
Y = np.cos(m_theta)
```

```python
dwin=dt
win=5*dt #Window
app_ref=20 #Aperture for conventional migration
app_dif=80 #Aperture for diffraction migration

gather1 = files[5] 
isx = 75
```

```python
with open('Data/CSG_diffbar/TTh_100x150_250mps.npy', 'rb') as f:
    TTh = np.load(f)

dt=0.004
timer=np.round(TTh/dt)+1
    
with open("Data/CSG_diffbar/raiosHomo_toohigh.npy", 'rb') as f:
    TTh_toohigh = np.load(f)
    
with open("Data/CSG_diffbar/raiosHomo_toolow.npy", 'rb') as f:
    TTh_toolow = np.load(f)

timer=np.round(TTh/dt)+1
timer_toohigh=np.round(TTh_toohigh/dt)+1
timer_toolow=np.round(TTh_toolow/dt)+1
```

```python
refl,diff = df.kirchhoffMigration(gather1,isx,dx,dz,dt,win,dwin,app_ref,app_dif,TTh,X,Y,sm)
refl_toohigh,diff_toohigh = df.kirchhoffMigration(gather1,isx,dx,dz,dt,win,dwin,app_ref,app_dif,TTh_toohigh,X,Y,sm)
refl_toolow,diff_toolow = df.kirchhoffMigration(gather1,isx,dx,dz,dt,win,dwin,app_ref,app_dif,TTh_toolow,X,Y,sm)

fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(13,4))

ax1.imshow(diff, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax1.set_title("Migration with Correct Velocity Model")
ax1.set_xlabel("Distance (m)")
ax1.set_ylabel("Depth (m)")

ax2.imshow(diff_toohigh, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax2.set_title("5% Higher Velocity Model")
ax2.set_xlabel("Distance (m)")
ax2.set_ylabel("Depth (m)")

ax3.imshow(diff_toolow, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax3.set_title("5% Lower Velocity Model")
ax3.set_xlabel("Distance (m)")
ax3.set_ylabel("Depth (m)")

plt.tight_layout()
plt.show()
```

```python
refls,diffs = df.kirchhoffMigration(files,isx,dx,dz,dt,win,dwin,app_ref,app_dif,TTh,X,Y,sm)
refls_toohigh,diffs_toohigh = df.kirchhoffMigration(files,isx,dx,dz,dt,win,dwin,app_ref,app_dif,TTh_toohigh,X,Y,sm)
refls_toolow,diffs_toolow = df.kirchhoffMigration(files,isx,dx,dz,dt,win,dwin,app_ref,app_dif,TTh_toolow,X,Y,sm)

fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(13,4))

ax1.imshow(diffs, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax1.set_title("Migration with Correct Velocity Model")
ax1.set_xlabel("Distance (m)")
ax1.set_ylabel("Depth (m)")

ax2.imshow(diffs_toohigh, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax2.set_title("5% Higher Velocity Model")
ax2.set_xlabel("Distance (m)")
ax2.set_ylabel("Depth (m)")

ax3.imshow(diffs_toolow, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax3.set_title("5% Lower Velocity Model")
ax3.set_xlabel("Distance (m)")
ax3.set_ylabel("Depth (m)")

plt.tight_layout()
plt.show()
```
