## Migração 

******

Abertura e janela

```python
def taper(ntr,ns,app,isx,igx):
    
    ar = np.zeros([ns,ntr])
    cmp = int((isx+igx)/2)
    window = np.hanning(2*app)
    
    if (cmp-app)<0: 
        if (cmp+app)>ntr:
            ntr_2 = int(ntr/2)
            lw_2 = int(len(window)/2)            
            ar[:,:] = window[(lw_2 - ntr_2):(lw_2 + ntr_2)]  
        else:
            ar[:,0:(cmp+app)] = window[abs(cmp-app):]   
    elif (cmp+app)>ntr:
        ar[:,(abs(cmp-app)):] = window[0:(ntr - abs(cmp-app))]
    else:
        ar[:,(cmp-app):(cmp+app)] = window 
    
    return ar
```

```python
#timer é a tabela de tempos gerada em raytracer, convertida pra índices
#isx é a coordenada do tiro no grid
#nx,nz número de amostras em x e z
#dt
#gather é o Common Shot Gather

app=15
win=10
dwin=0.1
        
        
def migvsp_winapp(timer,isx,dt,gather,win,dwin,app):
    window = np.arange(-win,win,dwin)
    [nt,ntr]=gather.shape
    [ntr2,nz,nx]=timer.shape
    if ntr!=ntr2:
        print('Gather e traveltime table tem numero diferente de traços')
        
    mig=np.zeros([nz,nx])
    
    IX = np.arange(0,nx*dx,dx)
    IZ = np.arange(0,nz*dz,dz)
    [IIX,IIZ] = np.meshgrid(IX,IZ)
    # Loop over each trace of the shot gather at src isx
    for igx in range(0,ntr):
        trace_win = np.zeros([nz,nx])
        R = np.sqrt(IIZ**2 + (IIX-(igx+isx)/2*dx)**2)
        r_mask = (R==0)
        R[r_mask]= dx/1000
        obli = IIZ/R
        trace_app = taper(ntr,nz,app,isx,igx) 
        
        #escrever aqui em cada entrada de obli é o valor do cos(theta(ix,iz,igx))
        
        for j in range(len(window)):
            t = timer[isx,0:nz,0:nx] + timer[igx,0:nz,0:nx] #t_{d}
            twin = t + window[j]
            t2 = (twin<nt)*twin 
            trace1=gather1h.T[np.ix_([igx],t2.flatten().astype(np.int32))] 
            trace1 = trace1.reshape([nz,nx])
            trace1 = trace1*trace_app
            trace_win = trace_win+trace1
        
        mig[0:nz,0:nx]=mig[0:nz,0:nx] + trace_win*obli
        
    return mig
```

```python
win=10*dt
dwin=dt
app=20

mig_winapp = migvsp_winapp(timer,isx,dt,gather,win,dwin,app)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(8,4))

ax1.imshow(mig_winapp, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax1.set_title("Imagem Migrada")
ax1.set_xlabel("Distancia (m)")
ax1.set_ylabel("Profundidade (m)")

ax2.imshow(R, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax2.set_title("Modelo Original")
ax2.set_xlabel("Distancia (m)")
ax2.set_ylabel("Profundidade (m)")

plt.tight_layout()
plt.show()
```

Definindo o Migrador para todos os tiros sintéticos

```python
def migstack_winapp(timer,isx,dt,files,win,dwin,app):
    
    migs = []
        
    for count,gather in enumerate(files):
        
        isx = count*5
        print(f"shot {isx}")
    
        window = np.arange(-win,win,dwin)
        [nt,ntr]=gather.shape
        [ntr2,nz,nx]=timer.shape
        if ntr!=ntr2:
            print('Gather e traveltime table tem numero diferente de traços')

        mig=np.zeros([nz,nx])
        mig_final=np.zeros([nz,nx])

        IX = np.arange(0,nx*dx,dx)
        IZ = np.arange(0,nz*dz,dz)
        [IIX,IIZ] = np.meshgrid(IX,IZ)
        # Loop over each trace of the shot gather at src isx
        for igx in range(0,ntr):
            trace_win = np.zeros([nz,nx])
            R = np.sqrt(IIZ**2 + (IIX-(igx+isx)/2*dx)**2)
            r_mask = (R==0)
            R[r_mask]= dx/1000
            obli = IIZ/R
            trace_app = taper(ntr,nz,app,isx,igx) 

            for j in range(len(window)):
                t = timer[isx,0:nz,0:nx] + timer[igx,0:nz,0:nx] #t_{d}
                twin = t + window[j]
                t2 = (twin<nt)*twin 
                trace1=gather.T[np.ix_([igx],t2.flatten().astype(np.int32))] 
                trace1 = trace1.reshape([nz,nx]) #seção? imshow em um separado
                trace1 = trace1*trace_app
                trace_win = trace_win+trace1

            mig[0:nz,0:nx]=mig[0:nz,0:nx] + trace_win*obli
        
        migs.append(mig)
    
    mig_final = np.add.reduce(migs)
        
    return mig_final
```

```python
win=10*dt
dwin=dt
app=20
isx=0

migstack = migstack_winapp(timer,isx,dt,files,win,dwin,app)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(8,4))

ax1.imshow(migstack, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax1.set_title("Imagem Migrada")
ax1.set_xlabel("Distancia (m)")
ax1.set_ylabel("Profundidade (m)")

ax2.imshow(R, aspect="auto",extent=[0,dx*nx,dz*nz,0])
ax2.set_title("Modelo Original")
ax2.set_xlabel("Distancia (m)")
ax2.set_ylabel("Profundidade (m)")

plt.tight_layout()
plt.show()
```

```python
#gH = np.gradient(timer, axis=2)
#gV = np.gradient(timer, axis=1)

gH = np.diff(timer, axis=2, append=1)
gV = np.diff(timer, axis=1, append=1)
def p_vector(gH, gV,n_horizontal, n_vertical, igx,isx):
        
    
    
    prV = gV[igx,:,:] 
    prH = gH[igx,:,:] 
    
    psV = gV[isx,:,:] 
    psH = gH[isx,:,:] 
    
    norma = np.array(np.sqrt((psH + prH)**2 + (psV + prV)**2))
    #Remover valores norma 0
    norma = np.where(norma>0.001,norma,0.01)
    
    pH = (psH + prH)/norma
    pV = (psV + prV)/norma
    
    w = (pH * n_horizontal) + (pV * n_vertical)
    #w = w>1-0.09
    
    return w
    
w = p_vector(gH, gV, n_horizontal, n_vertical, 55,55)
plt.imshow(w, aspect="auto")
#plt.xlim([40,60])
#plt.ylim([40,60])
plt.colorbar()
plt.show()
```

```python
def migvsp2(timer,isx,dt,gather):
    gH = np.diff(timer, axis=2, append=0)
    gV = np.diff(timer, axis=1, append=0)
    [nt,ntr]=gather.shape
    [ntr2,nz,nx]=timer.shape
    if ntr!=ntr2:
        print('Gather e traveltime table tem numero diferente de traços')
    #tmin=np.min(timer.flatten('F'))
    #if tmin==0:
    #    timer=timer+1
    mig=np.zeros([nz,nx])
    migD=np.zeros([nz,nx])
    # Loop over model
    
    for igx in range(0,ntr):
        w = p_vector(gH, gV, n_horizontal, n_vertical, igx,isx)
        for ix in range(0,nx):
            for iz in range(0,nz):
            # Loop over each trace of the shot gather at src isx
                if np.isnan(w[iz,ix]):
                    print('nan',iz,ix)
                t = timer[isx,iz,ix] + timer[igx,iz,ix]
                if t<nt:
                    mig[iz,ix] = mig[iz,ix]+w[iz,ix]*gather[t.astype(np.int32),igx]
                    migD[iz,ix] = migD[iz,ix]+((1-w[iz,ix])*gather[t.astype(np.int32),igx])
                    
    
    return mig,migD
```
